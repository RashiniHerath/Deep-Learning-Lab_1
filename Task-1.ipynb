```python
import numpy as np

array_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 Array from exponential distribution:\n", array_4x4)
```

```python
import matplotlib.pyplot as plt

# Create a 100000x1 array from an exponential distribution
array_100000 = np.random.exponential(scale=1.0, size=100000)

# Create a 100000x1 array from a uniform distribution
uniform_array = np.random.uniform(low=0.0, high=1.0, size=100000)

# Create a 100000x1 array from a normal distribution
normal_array = np.random.normal(loc=0.0, scale=1.0, size=100000)

# Plot the histograms
plt.hist(array_100000, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_array, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_array, density=True, bins=100, histtype="step", color="green", label="Normal")

plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create the x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = x**2 + y**2

# Plot the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_title("3D plot of Z = X^2 + Y^2")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

plt.show()
```

```python
import seaborn as sns

pearson_corr = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='pearson')

spearman_corr = df[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].corr(method='spearman')

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation')

plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Correlation')

plt.show()
```
